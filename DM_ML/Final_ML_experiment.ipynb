```python
all_models = models()
```

```python
results = []

for model_name in all_models.index:
    created_model = create_model(model_name)

    finalized_model = finalize_model(created_model)

    test_predicitons = predict_model(finalized_model, data=test_df)

    metrics = pull()
    metrics["Model"] = model_name
    results.append(metrics)
```

```python
results_df = pd.concat(results)
```
